```r
library ('tidyverse')
```

```r
plastics = readr::read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-01-26/plastics.csv')
```

```r
plastics |> head()
```

```r
options(repr.plot.height = 10, repr.plot.width = 10)
plastics |>
mutate(country = country
       |> fct_infreq() 
       |> fct_rev()) |>
ggplot(aes(y=country, fill= year)) + geom_bar() + facet_wrap(~year) + theme(axis.text.x = element_text(angle = 90, hjust = 1)) + labs(title = 'Bar Chart of Plastic Pollution Collected by Country 2019-2020') + labs(x = 'Grand Total Count of Plastics') + labs(y = 'Country') 
```

```r
options(repr.plot.width=20, repr.plot.height=5)
ggplot(plastics, aes(x = country, y = grand_total)) + geom_point(position = position_jitter(h = 0, w = 0.20), size = 1) + theme(axis.text.x = element_text(angle = 45, hjust = 1))  + labs(title = 'Scatter Plot of Plastic Pollution Collected by Country 2019-2020') + labs(x = 'Country') + labs(y = 'Amount of Plastic Collected') 
```

```r
options(repr.plot.width = 15, repr.plot.height = 6)
ggplot(plastics, aes(x = year, y = grand_total, color = country)) + geom_line() + scale_x_continuous(breaks=seq(2019, 2020, 1) ) + labs(title = 'Line Graph of Total Plastic Pollution Collected by Country 2019-2020') + labs(x = 'Country') + labs(y = 'Amount of Plastic Collected')
```
